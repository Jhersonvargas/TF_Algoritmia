```python
import pandas as pd
import random

# Generar datos aleatorios para productos
categorias = ["Herramientas", "Ferretería"]
productos = []
for i in range(1, 51):
    producto = {
        "ID Producto": i,
        "Nombre": f"Producto{i}",
        "Categoría": random.choice(categorias),
        "Precio": round(random.uniform(5, 100), 2)
    }
    productos.append(producto)

# Crear un DataFrame para los productos
df_productos = pd.DataFrame(productos)

# Generar datos aleatorios para registros de pedidos
registros_pedidos = []
for i in range(1, 51):
    productos_pedido = random.sample(list(df_productos["Nombre"]), random.randint(1, 5))
    total_pedido = sum(df_productos[df_productos["Nombre"].isin(productos_pedido)]["Precio"])
    pedido = {
        "ID Pedido": i,
        "Fecha de Pedido": f"2023-08-{str(i).zfill(2)}",
        "Cliente": f"Cliente{i}",
        "Productos Pedido": ", ".join(productos_pedido),
        "Total Pedido": round(total_pedido, 2)
    }
    registros_pedidos.append(pedido)

# Crear un DataFrame para los registros de pedidos
df_registros_pedidos = pd.DataFrame(registros_pedidos)

# Especificar la ruta y el nombre del archivo Excel
nombre_archivo = "BD-eltornillofeliz.xlsx"

# Crear un objeto ExcelWriter para escribir varios DataFrames en el mismo archivo
with pd.ExcelWriter(nombre_archivo) as writer:
    df_productos.to_excel(writer, sheet_name="Productos", index=False)
    df_registros_pedidos.to_excel(writer, sheet_name="RegistrosPedidos", index=False)

print(f"Base de datos para 'El Tornillo Feliz' con 50 registros guardada en {nombre_archivo}")

df_productos
```

```python
import pandas as pd
import math

# Cargar la base de datos desde el archivo Excel
nombre_archivo = "BD-eltornillofeliz.xlsx"
df_productos = pd.read_excel(nombre_archivo, sheet_name="Productos")
df_registros_pedidos = pd.read_excel(nombre_archivo, sheet_name="RegistrosPedidos")

# Función para calcular la entropía de un conjunto de datos
def calcular_entropia(datos):
    total = len(datos)
    categorias = datos["Categoría"].value_counts()
    entropia = 0
    for categoria_count in categorias:
        probabilidad = categoria_count / total
        entropia -= probabilidad * math.log2(probabilidad)
    return entropia

# Función para calcular la ganancia de información
def calcular_ganancia(datos, columna):
    entropia_inicial = calcular_entropia(datos)
    valores = datos[columna].unique()
    ganancia = entropia_inicial
    for valor in valores:
        subconjunto = datos[datos[columna] == valor]
        probabilidad = len(subconjunto) / len(datos)
        ganancia -= probabilidad * calcular_entropia(subconjunto)
    return ganancia

# Calcular la ganancia de información para la columna "Categoría"
ganancia_categoria = calcular_ganancia(df_productos, "Categoría")

print("Ganancia de Información para la columna 'Categoría':", ganancia_categoria)
```
