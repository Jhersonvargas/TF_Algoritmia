```python
class Usuario:
    def __init__(self, nombre, email, telefono):
        self.nombre = nombre
        self.email = email
        self.telefono = telefono

```

```python
class Producto:
    def __init__(self, id_producto, nombre, categoria, precio):
        self.id_producto = id_producto
        self.nombre = nombre
        self.categoria = categoria
        self.precio = precio

```

```python
def registrarse():
    nombre = input("Ingrese su nombre: ")
    email = input("Ingrese su email: ")
    telefono = input("Ingrese su número de teléfono: ")
    return Usuario(nombre, email, telefono)


```

```python
def hacer_pedido(catalogo_productos):
    print("Catálogo de productos:")
    for producto in catalogo_productos:
        print(f"{producto.id_producto}. {producto.nombre} - ${producto.precio}")
    
    productos_seleccionados = []
    while True:
        id_producto = int(input("Seleccione un producto por su ID (0 para finalizar): "))
        if id_producto == 0:
            break
        producto_seleccionado = next(p for p in catalogo_productos if p.id_producto == id_producto)
        productos_seleccionados.append(producto_seleccionado)
    
    return productos_seleccionados

```

```python
def anular_compra(pedido):
    anulable = True  # Supongamos que todos los pedidos son anulables por defecto
    for producto in pedido.productos:
        if producto.categoria == "Herramientas":
            anulable = False  # Si hay productos de la categoría "Herramientas", no se puede anular
            break
    
    if anulable:
        print("Pedido anulado. Se emitirá un reembolso.")
    else:
        print("No es posible anular este pedido.")

```

```python
def proceso_pedido(usuario, catalogo_productos):
    print(f"Bienvenido, {usuario.nombre}!")
    productos_seleccionados = hacer_pedido(catalogo_productos)
    pedido = Pedido(usuario, productos_seleccionados)

    return pedido

```

```python
class Pedido:
    def __init__(self, usuario, productos):
        self.usuario = usuario
        self.productos = productos

```

```python
def main():
    catalogo_productos = [
        Producto(1, "Martillo", "Herramientas", 20),
        Producto(2, "Tornillos", "Ferretería", 0.10),
        Producto(3, "Taladro Eléctrico", "Herramientas", 150),
    ]
    
    while True:
        print("Bienvenido a la Ferretería 'El Tornillo Feliz'")
        print("1. Registrarse")
        print("2. Hacer Pedido")
        print("3. Anular Compra")
        print("4. Salir")
        
        opcion = int(input("Seleccione una opción: "))
        
        if opcion == 1:
            usuario = registrarse()
            # Continuar con el proceso
            pedido = proceso_pedido(usuario, catalogo_productos)
        elif opcion == 2:
            if 'usuario' in locals():
                # Continuar con el proceso
                pedido = proceso_pedido(usuario, catalogo_productos)
            else:
                print("Primero debe registrarse antes de hacer un pedido.")
        elif opcion == 3:
            if 'pedido' in locals():
                anular_compra(pedido)
            else:
                print("No hay un pedido para anular.")
        elif opcion == 4:
            print("Gracias por visitar 'El Tornillo Feliz'. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción válida.")

if __name__ == "__main__":
    main()
```
